```python
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials, SpotifyOAuth

SPOTIPY_CLIENT_ID = "0be1da36f0594963b21a361ae5bd5d07"
SPOTIPY_CLIENT_SECRET = "093f6e330b414e4cbda05e95dbc0bef7"
SPOTIPY_REDIRECT_URI = 'http://localhost:8080/callback'

client_id = SPOTIPY_CLIENT_ID
client_secret = SPOTIPY_CLIENT_SECRET
redirect_uri = SPOTIPY_REDIRECT_URI


client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_global_top_50():
    # Global Top 50 Playlist URI (this URI is subject to change)
    playlist_uri = 'spotify:playlist:37i9dQZEVXbMDoHDwVN2tF'

    # Fetch playlist
    results = sp.playlist_tracks(playlist_uri)

    songs_data = []
    for idx, item in enumerate(results['items']):
        track = item['track']
        song_info = {
            'title': track['name'],
            'artist': track['artists'][0]['name'],
            'album': track['album']['name'],
            # Spotify doesn't provide streaming numbers via Spotipy, placeholder added
            'streaming_numbers': 'N/A',
            # Trending status requires historical data comparison, placeholder added
            'trending': 'N/A'
        }
        songs_data.append(song_info)

    # Convert list of dictionaries to DataFrame
    return pd.DataFrame(songs_data)

# Get the top 50 songs and convert to DataFrame
top_50_df = get_global_top_50()
print(top_50_df)
```

```python
# Initialize the Spotipy client
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id, client_secret=client_secret, redirect_uri=redirect_uri, scope='user-library-read'))

# Replace with the song and artist you want to search for
song_name = "All I Want for Christmas Is You"
artist_name = "Mariah Carey"

results = sp.search(q=f'track:{song_name} artist:{artist_name}', type='track')

# Extract the track URI and get the number of streams
if len(results['tracks']['items']) > 0:
    track_uri = results['tracks']['items'][0]['uri']
    audio_analysis = sp.audio_analysis(track_uri)
    num_streams = audio_analysis['track']['num_streams']
    print(f"The song '{song_name}' by '{artist_name}' has {num_streams} streams.")
else:
    print("Song not found.")
```
